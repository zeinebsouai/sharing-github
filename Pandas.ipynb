```python
import pandas as pd
import numpy as np
exam_data = {'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']}
df=pd.DataFrame(exam_data)
df= df.set_axis(list('abcdefghij'),axis=0)
print(df)
print(df.head(3))
df.dropna(axis=0,how='any',inplace=True)
print(df)
print(df[['name','score']])
dictio={'name' : ["Suresh"], 'score': [15.5], 'attempts': [1], 'qualify': ["yes"]}
data2=pd.DataFrame(dictio)
df=pd.concat([df,data2])
df.index.values[-1] = 'k'
print(df)
df.drop('attempts',axis=1,inplace=True)
print(df)
lis=[]
for x in df.iloc[:,1]:
    if x<10:
        lis.append(0)
    else:
        lis.append(1)
print(lis)
df2 = pd.DataFrame(lis, columns =['success'])
df2.index=df.index
df=pd.concat([df,df2],axis=1)
print(df)
df.to_csv('my_data')
```
